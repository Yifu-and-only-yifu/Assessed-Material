```python
import random
from collections import Counter
import matplotlib.pyplot as plt

def simulate_hat_process():
    slips = list(range(1, 2026))
    while len(slips) > 1:
        a, b = random.sample(slips, 2)
        slips.remove(a)
        slips.remove(b)
        diff = max(a, b) - min(a, b)
        slips.append(diff)
    return slips[0]

def run_experiments(n):
    results = [simulate_hat_process() for _ in range(n)]
    freq = Counter(results)
    return freq

def plot_results(freq):
    intervals = [(1, 100), (101, 200), (201, 300), (301, 400), (401, 500), (501, 600), (601, 700), (701, 800), (801, 900), (901, 1000), (1001, 1100), (1101, 1200), (1201, 1300), 
                 (1301, 2100), (2101, 3400)]
    interval_labels = [f"{low}-{high}" for low, high in intervals]
    interval_counts = []
    for low, high in intervals:
        count = sum(freq[num] for num in freq if low <= num <= high)
        interval_counts.append(count)
    
    plt.figure(figsize=(15, 7))
    plt.bar(interval_labels, interval_counts, color='skyblue', edgecolor='black')
    plt.xlabel('Final number intervals')
    plt.ylabel('Occurrence count')
    plt.title('Frequency statistics by interval')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    n = int(input("Number of experiment time n: "))
    freq = run_experiments(n)
    plot_results(freq)
```
