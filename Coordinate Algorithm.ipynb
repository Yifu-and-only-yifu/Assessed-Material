```python
import numpy as np
import matplotlib.pyplot as plt

def spiral_coords(n):
    if n == 1:
        return (0, 0)
    k = int(np.ceil((np.sqrt(n) - 1) / 2))
    t = 2 * k
    m = (2 * k + 1) ** 2
    d = m - n
    if d < t:
        return (k - d, -k)
    elif d < 2 * t:
        d_prime = d - t
        return (-k, -k + d_prime)
    elif d < 3 * t:
        d_double = d - 2 * t
        return (-k + d_double, k)
    else:
        d_triple = d - 3 * t
        return (k, k - d_triple)

points = [spiral_coords(n) for n in range(1, 26)]
x = [p[0] for p in points]
y = [p[1] for p in points]
plt.figure(figsize=(6,6))
plt.scatter(x, y, c='blue', s=100)
for i, (xi, yi) in enumerate(points, start=1):
    plt.text(xi + 0.1, yi + 0.1, str(i), fontsize=12, color='red')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

plt.grid(True)
plt.title("Spiral Coordinates for Numbers 1-25")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

```python
import numpy as np

def spiral_coords(n):
    if n == 1:
        return (0, 0)
    k = int(np.ceil((np.sqrt(n) - 1) / 2))
    t = 2 * k
    m = (2 * k + 1) ** 2
    d = m - n

    if d < t:
        return (k - d, -k)
    elif d < 2 * t:
        d_prime = d - t
        return (-k, -k + d_prime)
    elif d < 3 * t:
        d_double = d - 2 * t
        return (-k + d_double, k)
    else:
        d_triple = d - 3 * t
        return (k, k - d_triple)

def manhattan_distance(coord):
    return abs(coord[0]) + abs(coord[1])

try:
    n = int(input("Please enter a positive integer: "))
    if n < 1:
        print("Please enter a positive integer (greater than or equal to 1)!")
    else:
        coord = spiral_coords(n)
        md = manhattan_distance(coord)
        print(f"Number {n} in the Ulam spiral has coordinates: {coord}")
        print(f"The Manhattan distance from the origin is: {md}")
except ValueError:
    print("Invalid input, please enter a positive integer.")
```
