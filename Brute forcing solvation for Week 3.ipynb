```python
import numpy as np

rng = np.random.default_rng()

sticks = 10

print('|'*sticks, f'{sticks} left')

def game(myturn):
    global sticks
    if myturn == 1:
        x = int(input('Your turn: How many to remove (up to 3)?'))
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks == 1:
            print('You win!')
            return 0
        game(0)

    else:
        x = rng.integers(1, min(sticks,4))
        print(f"Bot's turn: {x} removed")
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks == 1:
            print('You lose!')
            return 0
        game(1)

myturn = 1 # 1 = player starts first, 0 = bot starts first

game(myturn)
```

```python
import itertools
import numpy as np

def count_singular_4x4_binary_matrices():
    count_singular = 0
    total = 0

    for comb in itertools.combinations(range(16), 8):
        mat = np.zeros((4, 4), dtype=float)
        for idx in comb:
            mat[idx // 4, idx % 4] = 1.0
        det_val = np.linalg.det(mat)
        if round(det_val) == 0:
            count_singular += 1
        total += 1
    
    return count_singular, total

if __name__ == "__main__":
    num_singular, num_total = count_singular_4x4_binary_matrices()
    print(f"The total number of matrix: {num_total}")
    print(f"The total number of matrix with determinant 0: {num_singular}")
```

```python
import itertools
from sympy import Matrix

def position_most_zero_in_singular_matrices():

    zero_count = [0] * 16
    singular_count = 0
    for comb in itertools.combinations(range(16), 8):

        mat_list = [0]*16
        for idx in comb:
            mat_list[idx] = 1
        
        mat = Matrix([
            mat_list[0:4],
            mat_list[4:8],
            mat_list[8:12],
            mat_list[12:16]
        ])


        if mat.det() == 0:
            singular_count += 1
            for pos in range(16):
                if mat_list[pos] == 0:
                    zero_count[pos] += 1

    max_zero = max(zero_count)
    max_positions = [i for i, c in enumerate(zero_count) if c == max_zero]
    
    return singular_count, zero_count, max_zero, max_positions

if __name__ == "__main__":
    sc, zc, mz, mp = position_most_zero_in_singular_matrices()
    print(f"The total number of matrix with determinant 0: {sc}")
    print(f"In the uninvertible matrix，the number of 0 appears in entries (index 0-15):\n{zc}")
    print(f"The Maximum number that 0 appears in same position: {mz}")
    for p in mp:
        r, c = divmod(p, 4)
        print(f"index {p} Row {r}, Column {c} Number of time = {mz}")
```

```python
import itertools
from sympy import Matrix

def position_most_zero_in_singular_matrices_3x3():
    zero_count = [0] * 9
    singular_count = 0
    for comb in itertools.combinations(range(9), 5):
        mat_list = [0]*9
        for idx in comb:
            mat_list[idx] = 1
        mat = Matrix([
            mat_list[0:3],
            mat_list[3:6],
            mat_list[6:9]
        ])

        if mat.det() == 0:
            singular_count += 1

            for i in range(9):
                if mat_list[i] == 0:
                    zero_count[i] += 1

    max_zero = max(zero_count)
    max_positions = [i for i, c in enumerate(zero_count) if c == max_zero]
    
    return singular_count, zero_count, max_zero, max_positions

if __name__ == "__main__":
    sc, zc, mz, mp = position_most_zero_in_singular_matrices_3x3()
    print(f"In 3×3(5*1,4*0) Matrix：")
    print(f"The total number of matrix with determinant 0: {sc}")
    print(f"In the uninvertible matrix，the number of 0 appears in entries (index 0-15):\n   {zc}")
    print(f"The Maximum number that 0 appears in same position: {mz}")
    for p in mp:
        r, c = divmod(p, 3)
        print(f" index {p} Row {r}, Column {c}")
```

```python
import itertools
from sympy import Matrix

def position_most_one_in_invertible_matrices_3x3():

    one_count = [0] * 9
    invertible_count = 0
    for comb in itertools.combinations(range(9), 5):
        mat_list = [0]*9
        for idx in comb:
            mat_list[idx] = 1
        
        mat = Matrix([
            mat_list[0:3],
            mat_list[3:6],
            mat_list[6:9]
        ])
        
        if mat.det() != 0:
            invertible_count += 1
            for i in range(9):
                if mat_list[i] == 1:
                    one_count[i] += 1
    max_one = max(one_count)
    max_positions = [i for i, c in enumerate(one_count) if c == max_one]
    
    return invertible_count, one_count, max_one, max_positions

if __name__ == "__main__":
    inv_count, ones_count, max_ones, max_pos = position_most_one_in_invertible_matrices_3x3()
    print(f"In 3×3(5*1,4*0) Matrix，Number of invertible matrix: {inv_count}")
    print(f"In the invertible matrix，the number of 1 appears in entries (index 0-8):\n{ones_count}")
    print(f"The Maximum number that 1 appears in same position: {max_ones}")
    print("The index of that positions:", max_pos)
    for p in max_pos:
        r, c = divmod(p, 3)
        print(f" index {p} Row {r}, Column {c}")
```

```python
import itertools
from sympy import Matrix

def position_most_one_in_invertible_matrices_5x5():

    total_positions = 25  
    num_ones = 13

    one_count = [0] * total_positions

    invertible_count = 0

    comb_iter = itertools.combinations(range(total_positions), num_ones)

    for comb in comb_iter:

        mat_list = [0]*total_positions
        for idx in comb:
            mat_list[idx] = 1

        mat = Matrix([
            mat_list[ 0: 5],
            mat_list[ 5:10],
            mat_list[10:15],
            mat_list[15:20],
            mat_list[20:25]
        ])

        det_val = mat.det()
        if det_val != 0:
            invertible_count += 1

            for i in range(total_positions):
                if mat_list[i] == 1:
                    one_count[i] += 1

    max_one = max(one_count)

    max_positions = [i for i, c in enumerate(one_count) if c == max_one]

    return invertible_count, one_count, max_one, max_positions

if __name__ == "__main__":
    inv_count, ones_count, max_ones, max_pos = position_most_one_in_invertible_matrices_5x5()

    print(f"In 5×5(13*1,12*0) Matrix：")
    print(f"  Number of invertible matrix: {inv_count}")
    print(f"  In the invertible matrix，the number of 1 appears in entries (index 0-24):\n    {ones_count}")
    print(f"  The Maximum number that 1 appears in same position: {max_ones}")
    print(f"  The index of that positions: {max_pos}")
    for p in max_pos:
        r, c = divmod(p, 5)
        print(f"Row {r}, Column {c}")
```
